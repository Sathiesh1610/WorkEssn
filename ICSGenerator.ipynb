```python
# Paste this into a local Python 3 file (e.g., make_ics.py) and run:  python make_ics.py
from datetime import datetime, timedelta

dates = [
    "1-Feb","2-Feb","3-Feb","4-Feb","5-Feb","6-Feb","7-Feb","8-Feb",
    "9-Feb","10-Feb","11-Feb","12-Feb","13-Feb","14-Feb","15-Feb","16-Feb",
    "17-Feb","18-Feb","19-Feb","20-Feb","21-Feb","22-Feb","23-Feb","24-Feb",
    "25-Feb","26-Feb","27-Feb","28-Feb"
]

names = [
    "Lakshmi Prasanna","Naveen Kumar","Abishek Selvaraj","Sathiesh M",
    "Tamanna Tyagi","Kundan Verma","Lalit Kumar","Anju Kurian",
    "Shashi Patil","Mitali Jha","Poojitha Sankrana","Chamundi P",
    "Diksha","Nirmalraj C","Hansika S"
]

shifts = {

    "Lakshmi Prasanna": [
        "OFF","OFF","S1","S1","S1","S1","S1","OFF",
        "OFF","S3","S3","S3","S3","S3","OFF","OFF",
        "S2","S2","S2","S2","S2","OFF","OFF","S1",
        "S1","S1","S1","S1"
    ],

    "Naveen Kumar": [
        "S2","S2","S2","S2","S2","OFF","OFF","S1",
        "S1","S1","S1","S1","OFF","OFF","S1","S1",
        "S1","S1","S1","OFF","OFF","S3","S3","S3",
        "S3","S3","OFF","OFF"
    ],

    "Abishek Selvaraj": [
        "OFF","OFF","S2","S2","S2","S2","S2","OFF",
        "OFF","S1","S1","S1","S1","S1","OFF","OFF",
        "S3","S3","S3","OFF","S3","S3","OFF","OFF",
        "S2","S2","S2","S2"
    ],

    "Sathiesh M": [
        "OFF","S1","S2","S2","S2","S2","OFF","OFF",
        "S3","S3","S3","S3","S3","OFF","OFF","S2",
        "S2","S2","S2","S1","OFF","OFF","S1","S1",
        "S1","S1","S1","OFF"
    ],

    "Tamanna Tyagi": [
        "S1","S1","S1","S1","S1","OFF","OFF","S2",
        "S2","S2","S2","S2","OFF","OFF","S3","S3",
        "S3","S3","S3","OFF","OFF","S2","S2","S2",
        "S2","S2","OFF","S1"
    ],

    "Kundan Verma": [
        "S3","S3","S3","S3","S3","OFF","OFF","S2",
        "S2","S2","L","OFF","OFF","S1","S1","S1",
        "S1","S1","OFF","OFF","S1","S2","S2","S2",
        "S2","OFF","OFF","S3"
    ],

    "Lalit Kumar": [
        "S3","S3","S3","S3","OFF","OFF","S2","L",
        "S2","S2","S2","OFF","OFF","S1","S1","S1",
        "S1","S1","OFF","OFF","S2","S2","S2","S2",
        "OFF","OFF","S3","S3"
    ],

    "Anju Kurian": [
        "S1","OFF","OFF","S2","S2","S2","S3","S3",
        "OFF","OFF","S3","S3","S3","S3","S3","OFF",
        "OFF","S2","S2","S2","S2","S2","OFF","OFF",
        "S1","S1","S1","S1"
    ],

    "Shashi Patil": [
        "S2","S2","S2","S2","OFF","OFF","S1","S1",
        "S1","S1","S1","OFF","OFF","S2","S2","S2",
        "S2","S2","OFF","OFF","S1","S1","S1","S1",
        "S1","OFF","OFF","S2"
    ],

    "Mitali Jha": [
        "OFF","OFF","S1","S1","S1","S1","OFF","OFF",
        "OFF","S2","S2","S2","S2","S2","S2","OFF",
        "OFF","S3","S3","S3","S3","OFF","OFF","S2",
        "S2","S2","S2","S2"
    ],

    "Poojitha Sankrana": [
        "OFF","S2","S2","S2","S2","S2","OFF","OFF",
        "S1","S1","S1","S1","S1","OFF","OFF","S3",
        "S3","S3","S3","S3","OFF","OFF","S2","S2",
        "S2","S2","S2","OFF"
    ],

    "Chamundi P": [
        "S1","S1","S1","S1","S1","OFF","OFF","OFF",
        "S2","S2","S2","S2","S2","OFF","OFF","S1",
        "S1","S1","S1","S1","OFF","OFF","S2","S2",
        "S2","S2","S2","OFF"
    ],

    "Diksha": [
        "S2","S2","S2","OFF","OFF","S3","S3","S3",
        "S3","S3","OFF","OFF","S2","L","L","S2",
        "S2","S2","S2","OFF","OFF","S1","S1","S1",
        "S1","S1","OFF","OFF"
    ],

    "Nirmalraj C": [
        "OFF","S2","S2","S2","S2","S2","OFF","OFF",
        "S2","S2","S2","S2","S2","EVE","OFF","S2",
        "S2","S2","S2","S2","OFF","OFF","S2","S2",
        "S2","S2","S2","EVE"
    ],

    "Hansika S": [
        "OFF","S2","S2","S2","S2","S2","EVE","OFF",
        "S2","L","S2","S2","S2","OFF","OFF","L",
        "S2","S2","S2","S2","EVE","OFF","S2","S2",
        "S2","S2","S2","OFF"
    ]
}




def dtfmt(dt): return dt.strftime("%Y%m%dT%H%M%S")
shift_time = {"S1": ("06:00", "15:30"), "S2": ("14:00", "23:30"), "S3": ("22:00", "07:30")}

lines = ["BEGIN:VCALENDAR","VERSION:2.0","PRODID:-//Shift Calendar//Sathiesh M//EN"]

for i, d in enumerate(dates):
    sath = shifts["Sathiesh M"][i]
    ymd = datetime.strptime(f"{d}-2026", "%d-%b-%Y")
    lines.append("BEGIN:VEVENT")
    if sath in shift_time:
        st, et = shift_time[sath]
        start_dt = datetime.strptime(f"{ymd.strftime('%Y-%m-%d')} {st}", "%Y-%m-%d %H:%M")
        end_dt = datetime.strptime(f"{ymd.strftime('%Y-%m-%d')} {et}", "%Y-%m-%d %H:%M")
        if sath == "S3":
            end_dt += timedelta(days=1)
        lines.append(f"DTSTART;TZID=Asia/Kolkata:{dtfmt(start_dt)}")
        lines.append(f"DTEND;TZID=Asia/Kolkata:{dtfmt(end_dt)}")
        lines.append(f"SUMMARY:{sath} ({st}–{et})")

        # --- description builder ---
        same, prev, nex = [], [], []

        for nm in names:
            if nm == "Sathiesh M":
                continue

            # same shift
            if shifts[nm][i] == sath:
                same.append(nm)

        # previous & next mapping
        if sath == "S1":
            # Previous = S3 previous day
            if i > 0:
                for nm in names:
                    if nm != "Sathiesh M" and shifts[nm][i-1] == "S3":
                        prev.append(nm)
            # Next = S2 same day
            for nm in names:
                if nm != "Sathiesh M" and shifts[nm][i] == "S2":
                    nex.append(nm)

        elif sath == "S2":
            # Previous = S1 same day
            for nm in names:
                if nm != "Sathiesh M" and shifts[nm][i] == "S1":
                    prev.append(nm)
            # Next = S3 same day
            for nm in names:
                if nm != "Sathiesh M" and shifts[nm][i] == "S3":
                    nex.append(nm)

        elif sath == "S3":
            # Previous = S2 same day
            for nm in names:
                if nm != "Sathiesh M" and shifts[nm][i] == "S2":
                    prev.append(nm)
            # Next = S1 next day
            if i < len(dates)-1:
                for nm in names:
                    if nm != "Sathiesh M" and shifts[nm][i+1] == "S1":
                        nex.append(nm)

        # build description
        desc = []
        desc.append("Colleagues in same shift:")
        for n in same: desc.append(f"- {n}")
        desc.append("")
        desc.append("Previous shift:")
        for n in prev: desc.append(f"- {n}")
        desc.append("")
        desc.append("Next shift:")
        for n in nex: desc.append(f"- {n}")

        lines.append("DESCRIPTION:" + "\\n".join(desc))

    else:
        start_date = ymd.strftime("%Y%m%d")
        end_date = (ymd + timedelta(days=1)).strftime("%Y%m%d")
        lines.append(f"DTSTART;VALUE=DATE:{start_date}")
        lines.append(f"DTEND;VALUE=DATE:{end_date}")
        lines.append("SUMMARY:" + ("Leave" if sath=="L" else "Off" if sath=="OFF" else sath))
        lines.append("DESCRIPTION:")
        # --- Add color for OFF ---
        if sath == "OFF":
            #lines.append("CATEGORIES:Off")
            lines.append("COLOR:#33B679")  # Standard Sage Green
            lines.append("X-GOOGLE-CALENDAR-COLOR:#33B679")  # Google Calendar
            #lines.append("X-APPLE-CALENDAR-COLOR:#00FF00")   # Apple Calendar
    lines.append("END:VEVENT")

lines.append("END:VCALENDAR")

ics_text = "\n".join(lines)
print(ics_text)
```
